### RESULTS
I carried out an exploratorty data analysis on Just Capital corporate performance scores with a shape of 444 by 50 records. I scliced the dataset into the Just 100 companies and companies not in the Just 100. I further categorized the components in the dateset based on their scoring discretions, "Higher is better" (i.e an higher score will have a great impact) and "Lower is better"(a lower score will have a great impact). The scoring discretion was gotten from the Metric legent documentation. The mean of each feature of the just 100 companies were furthe taken to identify which component has the greatest impact among the just 100 companies. From the analysis, it was observed that among the high scoring discretion component, the top scoring components were PAY(i.e Workers Pay and Benfits) with a mean score of 59.8 and TREAT( Workers Treatment) with a mean score of 63.6. On the other hand, among the low scoring discretion components CUST(i.e Customer Treatment)with a mean score of 49.9 and ENV(Environment Impact)with a mean score of 51.3 are the top components.

Thus this were the top components identified.

To futher establish the result, i carried out a *linear regression* on the dataset to predict the future just score of just companies. The results is as follows:

###### *Results for the High Scoring Discretion Components*
Based on the 2 scoring discretion of the features of the dataset, i perfomed a linear regresion on 6 features("PAY", "SUPPLY", "TREAT", 'PROD', "JOBS", "INVEST") with the high scoring discretion and used the weight score of each of the companies as target variable. The results were as follows given the values of other components(Picking the top 2 +ve associations):

1. A % increase in workers pay (PAY) is associated with 24.3% increase in a company's just weight score.
2. A % increase in workers treatment (TREAT) is associated with 23.6% increase in a company's just weight score.
This was analysed with a Root Mean Square Error(RMSE) of 0.96.

###### *On the other hand results for Low Scoring Discretion Components*
A linear regression was also performed on 4 features ('COMM', 'CUST', 'ENV', 'LEAD')with low scoring discretion, using the weighted score of each company as the target variable.  The results were as follows given the values of other components(Picking the Lowest 2 +ve associations):

1. A % increase in Customer treatment (CUST) is associated with 0.3% increase in a company's just weight score.
2. A % increase in Community wellbeing-charity (COMM) is associated with 4.3% increase in a company's just weight score.
This was analysed with a Root Mean Square Error(RMSE) of 2.63.

##### In view of the result analysed the Top 3 components are PAY- i.e Workers Pay and Benefits, TREAT- Workers treatment and CUST - Customer treatment.

#### Performing linear regression on the features to identify the impact on the companies just score
Note: For higher scoring discretion features a +ve coefficient iindicates an increase while a -ve coeff indicates a decrease

```python
# Define feature Matrix X and response factor Y
feature_cols =["PAY", "SUPPLY", 'JOBS', 'PROD', 'INVEST', "TREAT"]  #for higher scoring discretion features
X = just400[feature_cols]
y = just400["WGT_SCORE"]
X.head()
```

```python
# Checking the type of X and Shape
print(type(X))
print(X.shape)
```

```python
# split X and y into training and testing set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =1)
```

```python
# print shapes of train and test 
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Linear Regression in scikit learn

```python
# import model
from sklearn.linear_model import LinearRegression

#instantiate Linear Regression
linreg = LinearRegression()

# fit the model to training data(Learn the coefficients)
linreg.fit(X_train, y_train)

#create a python list of feature name
feature_cols =["PAY", "SUPPLY", "TREAT", 'PROD', "JOBS", "INVEST"]  

# use the list to select a subset of the original dataframe
X = just400[feature_cols]

# select a series from the dataframe
y = just400["WGT_SCORE"]

#split into training aand testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =1)

# fit the model to training data(Learn the coefficients)
linreg.fit(X_train, y_train)

# make predictions on testing set
y_pred = linreg.predict(X_test)

#compute the RMSE of our predictions
from sklearn import metrics
np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```

```python
# Intercepting model coefficient
print(linreg.intercept_)
print(linreg.coef_)
```

#### Performing linear regression on lower scoring discretion features to identify the impact on the companies just score

```python
#create a python list of feature name
feature_colsLs =["COMM", 'CUST','ENV', "LEAD"]  

# use the list to select a subset of the original dataframe
XLs = just400[feature_colsLs]

# select a series from the dataframe
yLs = just400["WGT_SCORE"]

#split into training aand testing set
XLs_train, XLs_test, yLs_train, yLs_test = train_test_split(XLs, yLs, random_state =1)

# fit the model to training data(Learn the coefficients)
linreg.fit(XLs_train, yLs_train)

# make predictions on testing set
yLs_pred = linreg.predict(XLs_test)

#compute the RMSE of our predictions
np.sqrt(metrics.mean_squared_error(yLs_test,yLs_pred))
```
